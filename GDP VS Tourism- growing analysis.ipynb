```python
data= df.copy()
```

```python
data.dtypes 
```

I need to pass tourism inbound and tourism outbound as numerical so i can calculate the mean of each region to fillna, since these columns are my targets for this analysis and the number of nan are a big part of the data.

So in this case i chose to fillna instead of dropna to keep the integrity of my data

<h3 style="color: #47FF64;"> Fillna() columns with the mean </h3>

```python
data["tourism_inbound"] = round(data["tourism_inbound"].str.replace("$", "").str.replace(",", ""),2)
```

i got an error when passing it as numeric due to the dollar sign, therefore i use ".replace" to clean it as well the commas
did round 2 because it is a big number and python would not read it all, also for my analysis is better

```python
data["tourism_outbound"] = round(data["tourism_outbound"].str.replace("$", "").str.replace(",", ""),2)
```

```python
data["tourism_inbound"]= data["tourism_inbound"].astype(float)
data["tourism_outbound"]= data["tourism_outbound"].astype(float)
```

now i can start calculating the mean for each region. after i will fillna the nan of each country with the region mean value

```python
data["tourism_inbound"] = data["tourism_inbound"].fillna(0)  #filling the nan with 0 to be able to calculate the mean without nan
```

```python
data["tourism_outbound"] = data["tourism_outbound"].fillna(0)
```

```python
data.dtypes #we can see now that the dtype has changed to float
```

```python
data.isna().sum() #the Nan are handled for tourism_inbound and tourism_outbound
```

```python
mean_by_region_inbound = data.groupby('region')['tourism_inbound'].transform('mean')
data.loc[data['tourism_inbound'] == 0, 'tourism_inbound'] = mean_by_region_inbound
data["tourism_inbound"] = data["tourism_inbound"].apply(lambda x: '{:.2f}'.format(x))
```

after calculating the mean for each region i will fill only the 0 values, is the same logic as fillna but i needed to replace nan with 0 to be able to transform it to a float variable. So now i fill the 0 --> NAN with the mean corresponding to each region where the country belongs to

```python
#then i will do the same for tourism_outbound
mean_by_region_outbound = data.groupby('region')['tourism_outbound'].transform('mean')
data.loc[data['tourism_outbound'] == 0, 'tourism_outbound'] = mean_by_region_outbound 
data["tourism_outbound"] = data["tourism_outbound"].apply(lambda x: '{:.2f}'.format(x))
```

- new discovery

**.transform method()** allows you to execute a function for each value of the DataFrame.

https://www.w3schools.com/python/pandas/ref_df_transform.asp

now i will clean the GDP column the same way as i did the tourism columns

```python
data["gdp"] = data["gdp"].str.replace("$", "").str.replace(",", "")
data["gdp"]= data["gdp"].astype(float)
data["gdp"] = data["gdp"].fillna(0)
mean_by_region_gdp = data.groupby('region')['gdp'].transform('mean')
data.loc[data['gdp'] == 0, 'gdp'] = mean_by_region_gdp
data["gdp"] = data["gdp"].apply(lambda x: '{:.2f}'.format(x)) #i got a big number that did not fit in the column so i had to format it with2 decimals
```

Drop some more unnecessary columns like population by age since i have the totalk population already, which is more usefull for my analysis

```python
data= data.drop(columns=["health_exp_%_gdp","health_exp/capita","infant_mortality_rate","life_expectancy_female","life_expectancy_male","number_of_records","birth_rate","population_0-14","population_15-64","population_65+","population_urban"])
```

cleaning the date column to display only the year

```python
data["year"]=data["year"].str[-4:]
```
